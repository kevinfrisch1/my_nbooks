```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
avocado_data = pd.read_csv("C:\\Users\\kevin\\Desktop\\avocado.csv")
```

```python
avocado_data.head()
```

```python
avocado_data.info()
```

```python
avocado_data['Date'] = pd.to_datetime(avocado_data['Date'])
```

```python
avocado_data.info()
```

```python
avocado_data.query('year == 2015')[['Date', 'Total Bags', 'Small Bags']].head()
```

```python
df=pd.melt(avocado_data, id_vars="Date",
                                  value_vars=["Total Bags", "Small Bags"],
                                  var_name='Bags', value_name='Count')
```

```python
df
```

```python
sns.relplot(data=df, kind="line", x="Date", y="Count", hue="Bags")
```

```python
bins = [0, 10842.77, 107431.24, 433251.71, 62505646.52]
```

```python
group_names = ['Poor', 'Modest', 'Good', 'Excellent']
```

```python
avocado_data['Sales Volume'] = pd.cut(avocado_data['Total Volume'], bins, labels=group_names)
```

```python
avocado_data.head()
```

```python
sns.countplot(x='year', data=avocado_data, hue='Sales Volume', hue_order=['Poor', 'Modest', 'Good', 'Excellent'])
```

```python
df=pd.melt(avocado_data, id_vars="year",
           value_vars=["Small Bags", "Large Bags", "XLarge Bags"],
           var_name='Bags', value_name='Count')
```

```python
df
```

```python
sns.relplot(data=df, kind="line", x="year", y="Count", hue="Bags")
```
